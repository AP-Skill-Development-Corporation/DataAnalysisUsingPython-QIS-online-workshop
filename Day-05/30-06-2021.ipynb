### Matplotlib

```python
import matplotlib.pyplot as plt
```

#### Boxplot
- It gives a summary of numerical data throug quantiles
    - Min
    - Inter Quantile Range(IQR)
        - Lower quantile (25%)
        - median (50%)
        - Upper quantile (75%)
    - Max

```python
import numpy as np
x = np.random.randint(2500, 25000, 1000)
x
```

```python
plt.boxplot(x)
plt.show()
```

```python
help(plt.boxplot)
```

```python
plt.boxplot(x, notch = True)
plt.show()
```

```python
cse_st_id = np.random.randint(0,5000,50)
eee_st_id = np.random.randint(1000,3500,500)
```

```python
plt.boxplot([cse_st_id,eee_st_id], labels = ['CSE', 'EEE'], 
            patch_artist = True, boxprops = dict(facecolor = 'm',color = 'r'))
plt.show()
```

### Histogram
- Frequency distribution of numerical data

```python
h = [0, 0.3, 0.3, 0.9, 1.6, 1.7, 3, 3.6, 3.8, 5, 5.1, 5.2]
bins = 4
plt.hist(h, bins, rwidth = 0.4)
```

### Categorical plot

```python
help(sns.catplot)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris, kind = 'box', palette = 'gnplot')
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris, kind = 'box', palette = 'gnuplot2')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
sns.catplot(x = 'sex', y = 'survived', data = titanic, kind = 'bar')
```

```python
sns.catplot(x = 'sex', y = 'survived', hue = 'pclass', data = titanic, kind = 'bar')
```

### Joint plot

```python
iris.head()
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = 'sepal_length',y = 'petal_length', data = iris)
```

```python
sns.jointplot(x = 'sepal_length',y = 'petal_length', data = iris, color = 'purple')
```

```python
sns.jointplot(x = 'sepal_length',y = 'petal_length', data = iris, kind = 'hex', color = 'pink')
```

```python
sns.jointplot(x = 'sepal_length',y = 'petal_length', data = iris, kind = 'kde', color = 'm')
## KDE --> Kernel density estimation
#### Kde estimates the probability density curve in 1 one or more dimensions of continous variables
```

### Pairplot
- Pairwise relationship between each column

```python
iris.head(2)
```

```python
sns.pairplot(iris)
## Diagonlly --> It is pairing with itself
```

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
sns.pairplot(iris, vars = ['sepal_length','petal_width'], hue = 'species', palette = 'coolwarm')
```

```python
sns.pairplot(iris, hue = 'species', markers = ['+','D','^'],palette = 'gist_ncar')
```
