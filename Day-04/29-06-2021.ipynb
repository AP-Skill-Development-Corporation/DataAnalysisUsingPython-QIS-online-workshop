#### Handling missing values
- Mainly 4 methods are there
    - **To identify missing values**
        1. isnull() : Presence of missing values
        2. notnull() : Opposite to isnull()
        
    - **To manipulate null values**
        1. dropna() : Removes null values
        2. fillna() : Replace null values 

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Global%20Dataset/Market_Fact.csv')
```

### Replace null values using fillna()
- If data is numerical datatype, thenreplace the null value either with 0 or mean or mediun or any number
- If data categorical or object datatype, then replace the null value with most occured value

```python
df2 = df.copy()
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
## find the mean value
df2['Product_Base_Margin'].mean()
```

```python
df2['Product_Base_Margin'] = df2['Product_Base_Margin'].fillna(df2['Product_Base_Margin'].mean())
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
x = np.linspace(0,5,20)
x
```
