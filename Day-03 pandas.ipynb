## Creating pandas  Series by using python objcets

```python
t = (1,4,5,76,8)

s1 = pd.Series(t)
s1
```

```python
# To define our Own index lables, we need to access index attriute in pandas Series method
data = ['masthanvaali', 'venkatesh', 'dinesh', 'subhash','navya', 'sai teja']

name_saries = pd.Series(data, index=['M', 'V', 'D', "S", 'N', 'T'])

name_saries
```

```python
# the index attritute elemnets length must be equal to the length of input data 
data = ['masthanvaali', 'venkatesh', 'dinesh', 'subhash','navya', 'sai teja']

name_saries = pd.Series(data, index=['M', 'V', 'D', "S", 'N'])

name_saries
```

```python
# create a Series using a python dictionary object

dict_data = {i:ord(i) for i in "ABCDEF"}
print('Dictionary Data: ',dict_data)

s_dict = pd.Series(dict_data)

s_dict
```

```python
# create a pandas Series by using floationg point numbers as a values

import numpy as np 

float_data = np.random.rand(10)
print(float_data)

s_float = pd.Series(float_data, index=range(100, 100+len(float_data)))
s_float
```

```python
dict_data2 = {'name': 'Durga', 
             'College': "Qis",
             515: 'ROll number'}

s_dict2 = pd.Series(dict_data2)
s_dict2
```

```python
s_dict2.index = ['Name', 'College', "Roll_num"]
```

```python
s_dict2
```

```python
# to get only the values present in the series object we can use (.values attribute)
s_dict2.values
```

```python
s_float.values
```

```python
# create a pandas Series object with Dates
dates = pd.date_range(start= '2021-01',periods=10, freq='D') 
print('Dates: ',dates)

s_dates = pd.Series(dates)
s_dates
```

```python
date_month = pd.date_range(start='2021-07', periods=10, freq='M')

print(date_month)

s_dates2 = pd.Series(date_month.month_name(), index= date_month)
s_dates2
```

## Data Frames
- It a 2-Dimentional data Structure, 
- Very similar to tablar data's(sheets, dataBases etc,.)
- It have both column and row lables 
- It can store any type of data


### Create our own pandas DataFrame

```python
# create by using python list of list

data_list_of_list = [['Revathi', "mounika",'Dhatri'],[526,1,563]]

df1 = pd.DataFrame(data_list_of_list)
df1
```

```python
# create by using python list of list

data_list_of_list = [['Revathi', "mounika",'Dhatri'],[526,1,563]]

df1 = pd.DataFrame(data_list_of_list, index=['name','roll_num'])
df1
```

```python
data_array = np.arange(1,100)

df2 = pd.DataFrame(data_array)
df2
```

```python
df3 = pd.DataFrame(data_array*2)
df3
```
